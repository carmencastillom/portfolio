This game is based on the movie called 'Charlie and The Chocolate Factory.' You will play the role of Charlie during the game. Charlie is one of those who win a Golden Pass, and in comparison with the other four individuals, his family does not have many resources.

Your mission is to help Charlie successfully pass the three stages of the game so that, in the end, he can win the game and gain control of the Chocolate Factory for his use! To successfully obtain a Golden Ticket, Charlie (you) needs to pass the three stages that are explained below:

    1. Win a ticket.
    2. Pass one of the proofs included in the game.
    3. Make the final decision based on what Willy Wonka is to offering him.

```python
# defining a fail function with a message for the user.
def fail():
    print("Oops, you failed!")

# defining a win function with a message for the user. 

def win():
    print("Congratulations! You have won the game. The factory chocolate is now yours. "
          "I hope you enjoyed the chocolate world and living in such a dream place.")
    
# defining a get_user_funtion for prompt and user options. 

def get_user_choice(prompt, options):
    while True:
        user_input = input(prompt).lower()
        if user_input in options:
            return user_input
        else:
            print("Invalid option. Please choose a valid option.")

# printing a message to the user.
print(f"""
{'#' * 100}

INSTRUCTIONS:
Are you excited to dive into the Chocolate Word? I hope the answer is YES! 
You can make this possible while playing this game based on the movie Charlie
and the Chocolate Factory. 
During the game you are going to take Charlie´s Role! The goal is to pass the three stages so at the end game you will be able to get the Chocolate Factory for your use!


{'#' * 100}\n""")

input("Press < Enter > to begin.")            

# defining the function play_game which is the starting point of the game (first stage)

def play_game():
    
    print('#' * 100)
    print("\nWelcome to the first stage of the game!")
    print("In this part, you need to win a Golden Ticket Pass which will allow you to enter the factory. Let´s go 🍫")
    print('#' * 100)

    while True:
       
        choice_money = get_user_choice( 
            'You are walking on the street and you see money on the ground. '
            'What do you want to do? Type "grab" or "not grab"\n', 
            ["grab", "not grab"]
        )
       
        if choice_money == "grab":
            # Second Decision
            choice_shop = get_user_choice(
                'You are at the shop close to you. They are selling Willy Wonka Chocolate. '
                'Make a decision: Choose "buy" to get the chocolate or "not buy" to save the money for your family\n',
                ["buy", "not buy"]
            )

            if choice_shop == "buy":
                print('Congrats! You have won a Golden Pass')
                # Person list
             
                choice_person = ["Grandfather", "Grandmother", "Father", "Mother"]
                print("Select who is going to visit with you the factory. Hint: Old male family member")
                for index, person in enumerate(choice_person, 1):
                    print(f"{index}. {person}")

                choice = get_user_choice("Please, choose the number corresponding to the person\n", ["1", "2", "3", "4"])

                if choice == "1":
                    print("Lovely, your Grandfather is visiting with you the factory!")
                    print('Congratulations! You can go ahead to the second part of the game. Best of luck\n.')
                    input('Please <enter> to continue')
                    stage_2()
                    continue
                else:
                    print("This person is not able to go with you. Please, read the hint carefully and try again. Hint: The person who is going with you is an old male person")
                    fail()
                    
                    
                    

            else:
                # Logic for not buying
                print('You chose not to buy. Maybe next time.')
                fail() 
                
                play_again = input('Do you want to play again? Type "yes" or "no"\n').lower()
                if play_again == "yes":
                    continue
                else:
                    print("Thanks for playing. Goodbye!")
                    break

        else:
            # printing message for the user when not grabbing the money
            print("You chose not to grab the money. Maybe next time.")
            fail()
            
            play_again = input('Do you want to play again? Type "yes" or "no"\n').lower()
            if play_again == "yes":
                continue
            else:
                print("Thanks for playing. Goodbye!")
                break
                
# Defining the 2nd Stage 
def stage_2():
    
#Printing a Welcome to the 2nd Stage of the game   
    print('#' * 100)
    print("\nYou are now on the second stage of the game!")
    print("In this part, you need to win one of the tests Willy Wonka has prepared.")
    print("In this set up, there is a television in which any object that goes inside it gets smaller")
    print('#' * 100)


# List for the options on the second part of the game
    valid_options = ["go", "keep", "safe"]

# While Loop to try again the game while true
    while True:
        print('What should you do? Type the correct option:')
        for option in valid_options:
            print(f'"{option}" to {option.capitalize()}')

        television_proof = get_user_choice("Your choice: ", valid_options)

        # if/elif/els
        if television_proof == "go":
            print("Bad decision! You have died entering on the television")
            fail() #The user has failed the game
            play_again = input('Do you want to try again? Type "yes" or "no"\n').lower() #Presenting to the user the option to play again the game
            # if true start the 2nd stage again
            if play_again == "yes":
                continue
            else:
                print("Thanks for playing. Goodbye!")
                break
        elif television_proof == "keep":
            print("Well done! You have passed the test and will now proceed to the third stage!")
            input('Click here to go to the last part of the game!')
            stage_3()  # Starting the 3rd stage
            break
        elif television_proof == "safe":
            print("Bad decision! You should not have saved Charliemate as he would have died")
            fail()
            play_again = input('Do you want to try again? Type "yes" or "no"\n').lower()
            if play_again == "yes":
                continue
            else:
                print("Thanks for playing. Goodbye!")
                break

# Defining the 3rd Stage and the last one
def stage_3():
    # Printing a welcome to the Third Stage
    print('#' * 100)
    print("\nWelcome to the third stage of the game! Implement your logic here.!")
    print("In this part, you need to make an important decision.")
    print("Do not forget that you love too much your family so that is going to be an important factor in your decision!")
    print('#' * 100)
   
    # List elevator_options with the options to do
    elevator_options = ["Go Into", "Stay Aside"]
    
    print('You are about to take an elevator that gets you out of the factory. '
          'What do you want to do? Please type "Go Into" if you want to go into the elevator '
          'or type "Stay Aside" if you do not want to get into the elevator')
    
    # For loop to create an index on list so it is simpler to the user
    for index, elevator in enumerate(elevator_options, 1):
        print(f"{index}. {elevator}")

    while True:
        elevator_choice = get_user_choice("Please, choose the number of the decision that fits you better\n", ["1", "2"])
        
        #if/else 
        if elevator_choice == "1":
            print('Nice decision! You are about to finish the game very successfully.')
            break  # Exit the loop if the user makes the correct choice
        else:
            print('Incorrect! That was not the right decision. Please, try again to be able to finish the game.')
            fail()

         
    # List willy_wonka_options so the user can pick up an option
    willy_wonka_options = ["Suggest option", "Do not get the Chocolate Factory", "Get the factory"]
    
    # Printing an explanation for the user and the options to pick
    print('Willy Wonka has offered the Chocolate Factory for your use! However, you love your family so much, he does not allow you to go with your family. '
          'What do you want to do? Please type "Suggest option" if you want to insist on the idea of living with your family at the Chocolate Factory, '
          'type "Do not get the Chocolate Factory" if you do not want to get the Chocolate Factory, '
          'or type "Get the factory" if you want to live there without including your family.')
    
    #For while to get an index on the list willy_wonka_options
    for index, willy_wonka in enumerate(willy_wonka_options, 1):
        print(f"{index}. {willy_wonka}")

    willy_wonka_choice = get_user_choice("Please, choose the number corresponding to your decision\n", ["1", "2", "3"])
    
    # if/elif/else options 
    
    if willy_wonka_choice == "1":
        print("You have suggested a correct answer, and Willy Wonka agrees. You'll live with your family at the Chocolate Factory.")
        win() #Calling win funtion as the user has selected all the correct answer
        # Handle further game logic or end the game based on your requirements
        
    elif willy_wonka_choice == "2":
        print("You choose not to get the Chocolate Factory so you have lost the opportunity of getting the Chocolate Factory.")
        fail() #Calling Fail Function
        
    elif willy_wonka_choice == "3":
        print("You decide to get the Chocolate Factory without including your family and it is not the correct answer as Charlie loves too much his family that he would not give them along")
        fail() #Calling a Fail Function

    else:
        print("Invalid choice. Please, try again to continue the game.")

#Starting the game
play_game()




```
